```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

# Load the dataset
data = pd.read_csv('creditcard.csv')

# Data Exploration
print(data.head())  # View the first few rows of the dataset
print(data.info())  # Get information about the dataset

# Data Cleaning
# Check for missing values (if any)
print(data.isnull().sum())

# Handling outliers (example using z-score)
z_scores = (data['Amount'] - data['Amount'].mean()) / data['Amount'].std()
outlier_threshold = 3  # Set the z-score threshold for outliers
outliers = data[z_scores > outlier_threshold]
data = data[z_scores <= outlier_threshold]

# Handling class imbalance
print(data['Class'].value_counts())  # Check the class distribution
oversampler = SMOTE(random_state=42)
X_resampled, y_resampled = oversampler.fit_resample(data.drop('Class', axis=1), data['Class'])
data_resampled = pd.concat([pd.DataFrame(X_resampled), pd.Series(y_resampled, name='Class')], axis=1)
print(data_resampled['Class'].value_counts())  # Check the resampled class distribution

# Feature Scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data_resampled.drop('Class', axis=1))
data_scaled = pd.concat([pd.DataFrame(scaled_features, columns=data_resampled.drop('Class', axis=1).columns),
                         data_resampled['Class']], axis=1)

# Feature Engineering (if applicable)
# You can apply additional feature engineering techniques here

# Save the cleaned and preprocessed data
data_scaled.to_csv('creditcard_cleaned.csv', index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

```python
data = pd.read_csv('creditcard.csv')
```

```python
print(data.head())  # View the first few rows of the dataset
print(data.info())  # Get information about the dataset
```

```python
z_scores = (data['Amount'] - data['Amount'].mean()) / data['Amount'].std()
outlier_threshold = 3  # Set the z-score threshold for outliers
outliers = data[z_scores > outlier_threshold]
data = data[z_scores <= outlier_threshold]
```

```python
# Handling class imbalance
print(data['Class'].value_counts())  # Check the class distribution
oversampler = SMOTE(random_state=42)
X_resampled, y_resampled = oversampler.fit_resample(data.drop('Class', axis=1), data['Class'])
data_resampled = pd.concat([pd.DataFrame(X_resampled), pd.Series(y_resampled, name='Class')], axis=1)
print(data_resampled['Class'].value_counts())  # Check the resampled class distribution
```

```python
# Feature Scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data_resampled.drop('Class', axis=1))
data_scaled = pd.concat([pd.DataFrame(scaled_features, columns=data_resampled.drop('Class', axis=1).columns),
                         data_resampled['Class']], axis=1)
```

```python
# Save the cleaned and preprocessed data
data_scaled.to_csv('creditcard_cleaned.csv', index=False)
```

```python
from sklearn.cluster import DBSCAN

# Training DBSCAN clustering model
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(data_scaled.drop('Class', axis=1))
```

```python
from sklearn.mixture import GaussianMixture

# Training Gaussian Mixture Model clustering
gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(data_scaled.drop('Class', axis=1))
```

```python
from sklearn.cluster import AgglomerativeClustering

# Training Agglomerative Clustering model
agg_clustering = AgglomerativeClustering(n_clusters=2)
agg_clustering.fit(data_scaled.drop('Class', axis=1))
```

```python
from sklearn.cluster import KMeans

# Training K-means clustering model
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(data_scaled.drop('Class', axis=1))
```

```python
kmeans_predictions = kmeans.predict(data_scaled.drop('Class', axis=1))
kmeans_predictions
```

```python
dbscan_predictions = dbscan.labels_
dbscan_predictions
```

```python
gmm_predictions = gmm.predict(data_scaled.drop('Class', axis=1))
gmm_predictions
```

```python
agg_clustering_predictions = agg_clustering.labels_
agg_clustering_predictions
```

```python
from sklearn.decomposition import PCA

# Instantiate PCA
pca = PCA(n_components=2, random_state=42)

# Fit and transform the data
pca_features = pca.fit_transform(data_scaled.drop('Class', axis=1))

# Create a new DataFrame with the principal components
pca_data = pd.DataFrame(data=pca_features, columns=['PC1', 'PC2'])
pca_data['Class'] = data_scaled['Class']
```

```python
# Apply dimensionality reduction using PCA
pca_output = pca.transform(data_scaled.drop('Class', axis=1))
pca_output
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Given data
data = {
    'Item': ['Item 1', 'Item 2', 'Item 3', 'Item 4'],
    'Variable X': [50, 20, 70, 90],
    'Variable Y': [100, 115, 120, 200],
    'Variable Z': [2.0, 1.5, 3.0, 4.0]
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Normalize the data
for col in ['Variable X', 'Variable Y', 'Variable Z']:
    df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

# Plotting
plt.figure(figsize=(8, 6))
pd.plotting.parallel_coordinates(df, 'Item', colormap='viridis')

# Add labels
plt.title('Parallel Coordinates Plot')
plt.xlabel('Variables')
plt.ylabel('Normalized Values')

# Show the plot
plt.show()
```
